```python
from bs4 import BeautifulSoup
import requests
import numpy as np
import pandas as pd
page = requests.get("http://www.myweather2.com/City-Town/Malaysia/Petaling-Jaya/14-Day-Forecast.aspx")
page.content
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
tables = soup.find('table',id="ctl00_MiddleContentHolder_grdForecast_CityTown")
arrays = []
number = 1
array = []
```

```python
for table in tables.find_all('tr'):
    columns = table.find_all('td')
    for column in columns:
        #if ((column.get_text() == '') && !(column.find('img') is None))
        columnspacing = column.get_text().replace(u'%', u'% ').replace(u'  ', u'').replace(u'\xa0mm', u'').replace(u'\t', u'').replace(u'\n', u'').replace(u'\r', u'').replace(u'\xa0mb', u'').replace(u'\xa0km/h', u'')
        array.append(columnspacing)
        if number == 10:
            array[-1] = column.find('img').get('title')
            arrays.append(array)
            array = []
            number = 0
        number+=1
```

```python
arrays = np.array(arrays)
dataset = pd.DataFrame({'Day': arrays[:, 0], 'Local Time': arrays[:, 1], 'Dir': arrays[:, 2],
                       'Speed/Gust (km/h)': arrays[:, 3], 'Temp/Feels Like': arrays[:, 4], 'Humidity': arrays[:, 5],
                       'Pressure (mb)': arrays[:, 6], 'Cloud Amount': arrays[:, 7], 'Precip Amount': arrays[:, 8],
                       'Weather': arrays[:, 9]})
dataset
```
